```python
#Q1.Create a Line Chart to display the monthly sales of a company for 12 months:
# Import matplotlib library
import matplotlib.pyplot as plt

# Months data (X-axis)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Sales data (Y-axis) in thousands of ₹
sales = [50, 55, 60, 65, 70, 75, 80, 85, 78, 90, 95, 100]

# Create line chart
plt.plot(months, sales)

# Add title
plt.title("Monthly Sales of a Company")

# Label X-axis
plt.xlabel("Months")

# Label Y-axis
plt.ylabel("Sales (₹ in thousands)")

# Add grid for better readability
plt.grid()

# Display the chart
plt.show()
```

```python

#Q2.Plot a Multiple Line Chart showing the sales of three products (Product A, B, and C) for six months.
# Import matplotlib library
import matplotlib.pyplot as plt

# Months data (X-axis)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']

# Sales data for three products
product_A = [20, 25, 30, 35, 40, 45]
product_B = [15, 20, 25, 30, 35, 38]
product_C = [10, 15, 20, 25, 30, 35]

# Plot lines for each product
plt.plot(months, product_A, label='Product A')
plt.plot(months, product_B, label='Product B')
plt.plot(months, product_C, label='Product C')

# Add title
plt.title("Sales of Products A, B, and C")

# Label axes
plt.xlabel("Months")
plt.ylabel("Sales (₹ in thousands)")

# Show legend to differentiate lines
plt.legend()

# Add grid
plt.grid()

# Display the chart
plt.show()
```

```python
#Q4.Generate a Horizontal Bar Chart to visualize the population of five different cities:
# Import matplotlib library
import matplotlib.pyplot as plt

# City names (Y-axis)
cities = ['Delhi', 'Mumbai', 'Bangalore', 'Chennai', 'Kolkata']

# Population data (in lakhs)
population = [190, 200, 120, 110, 140]

# Create horizontal bar chart
plt.barh(cities, population)

# Add title
plt.title("Population of Different Cities")

# Label X-axis
plt.xlabel("Population (in lakhs)")

# Label Y-axis
plt.ylabel("Cities")

# Add grid for better readability
plt.grid(axis='x')

# Display the chart
plt.show()
```

```python
#Q5.Create a Stacked Bar Chart showing the number of male and female students in four departments:
# Import matplotlib library
import matplotlib.pyplot as plt

# Departments (X-axis)
departments = ['CSE', 'IT', 'ECE', 'ME']

# Number of male students
male = [40, 30, 35, 25]

# Number of female students
female = [35, 25, 30, 20]

# Create stacked bar chart
# 'bottom' parameter stacks female bars on top of male bars
plt.bar(departments, male, label='Male')
plt.bar(departments, female, bottom=male, label='Female')

# Add title
plt.title("Male and Female Students in Departments")

# Label axes
plt.xlabel("Departments")
plt.ylabel("Number of Students")

# Show legend
plt.legend()

# Add grid
plt.grid(axis='y')

# Display the chart
plt.show()
```

```python
#Q7.Create an Exploded Pie Chart showing daily time spent on activities like study, sleep, exercise, and entertainment:
# Import matplotlib library
import matplotlib.pyplot as plt

# Activities
activities = ['Study', 'Sleep', 'Exercise', 'Entertainment']

# Time spent (in hours)
hours = [6, 8, 2, 4]

# Explode effect (separates slices slightly)
explode = [0.1, 0, 0, 0]  # Only 'Study' slice is exploded

# Create pie chart
plt.pie(hours, labels=activities, autopct='%1.1f%%', explode=explode)

# Add title
plt.title("Daily Time Spent on Activities")

# Display the chart
plt.show()
```

```python
#Q8.Plot a Histogram to visualize the distribution of marks obtained by 100 students:
# Import matplotlib library
import matplotlib.pyplot as plt
import random

# Generate marks for 100 students (random values between 40 and 100)
marks = [random.randint(40, 100) for i in range(100)]

# Create histogram
# bins=10 divides marks into 10 groups
plt.hist(marks, bins=10)

# Add title
plt.title("Distribution of Marks of 100 Students")

# Label X-axis
plt.xlabel("Marks")

# Label Y-axis
plt.ylabel("Number of Students")

# Add grid for better understanding
plt.grid()

# Display the histogram
plt.show()
```

```python
#Q10.Develop a Bubble Chart showing population vs area of cities, where bubble size represents GDP:
# Import matplotlib library
import matplotlib.pyplot as plt

# Data
cities = ['Delhi', 'Mumbai', 'Bangalore', 'Chennai', 'Kolkata']

# X-axis: Population (in millions)
population = [19, 20, 12, 11, 14]

# Y-axis: Area (in sq km)
area = [1484, 603, 741, 426, 206]

# Bubble size: GDP (scaled for better visibility)
gdp = [300, 400, 150, 120, 180]
bubble_size = [g * 5 for g in gdp]  # scale sizes

# Create scatter plot (bubble chart)
plt.scatter(population, area, s=bubble_size)

# Add city names as labels near points
for i in range(len(cities)):
    plt.text(population[i], area[i], cities[i])

# Add title
plt.title("Population vs Area (Bubble Size = GDP)")

# Label axes
plt.xlabel("Population (millions)")
plt.ylabel("Area (sq km)")

# Add grid
plt.grid()

# Display the chart
plt.show()
```

```python
#Q11.Plot an Area Chart showing the yearly profit growth of a company over 6 years:
# Import matplotlib library
import matplotlib.pyplot as plt

# Years (X-axis)
years = [2018, 2019, 2020, 2021, 2022, 2023]

# Profit data (Y-axis) in lakhs
profit = [20, 25, 30, 35, 45, 55]

# Create area chart using fill_between
plt.fill_between(years, profit)

# Add title
plt.title("Yearly Profit Growth of a Company")

# Label axes
plt.xlabel("Year")
plt.ylabel("Profit (₹ in lakhs)")

# Add grid
plt.grid()

# Display the chart
plt.show()
```

```python
#Q12.Create a Step Chart to display electricity consumption during different time intervals in a day:
# Import matplotlib library
import matplotlib.pyplot as plt

# Time intervals (X-axis)
time = ['6 AM', '9 AM', '12 PM', '3 PM', '6 PM', '9 PM', '12 AM']

# Electricity consumption (Y-axis) in kWh
consumption = [2, 4, 6, 5, 7, 6, 3]

# Create step chart
plt.step(time, consumption)

# Add title
plt.title("Electricity Consumption Throughout the Day")

# Label axes
plt.xlabel("Time Intervals")
plt.ylabel("Consumption (kWh)")

# Add grid
plt.grid()

# Display the chart
plt.show()
```

```python
#Q13.Generate a Box Plot to show the salary distribution of employees in three different departments:
# Import matplotlib library
import matplotlib.pyplot as plt

# Salary data for each department (in ₹ thousands)
hr = [20, 25, 30, 35, 40]
it = [30, 35, 40, 45, 50]
sales = [25, 30, 35, 40, 45]

# Combine all data into a list
data = [hr, it, sales]

# Create box plot
plt.boxplot(data)

# Add labels for each department
plt.xticks([1, 2, 3], ['HR', 'IT', 'Sales'])

# Add title
plt.title("Salary Distribution by Department")

# Label Y-axis
plt.ylabel("Salary (₹ in thousands)")

# Add grid
plt.grid()

# Display the plot
plt.show()
```

```python

#Q14.Create a Violin Plot to visualize the distribution of exam scores in different classes:
# Import matplotlib library
import matplotlib.pyplot as plt

# Exam scores for different classes
class_A = [45, 50, 55, 60, 65, 70, 75]
class_B = [50, 55, 60, 65, 70, 75, 80]
class_C = [40, 45, 50, 55, 60, 65, 70]

# Combine data
data = [class_A, class_B, class_C]

# Create violin plot
plt.violinplot(data)

# Add labels for each class
plt.xticks([1, 2, 3], ['Class A', 'Class B', 'Class C'])

# Add title
plt.title("Exam Score Distribution in Different Classes")

# Label Y-axis
plt.ylabel("Scores")

# Add grid
plt.grid()

# Display the plot
plt.show()
```

```python

#Create Subplots in a single figure that display four charts: Line Chart, Bar Chart, Pie Chart, and Scatter Plot:
# Import matplotlib library
import matplotlib.pyplot as plt

# Create a figure with 2 rows and 2 columns of subplots
plt.figure()

# --------------------------------
# 1. Line Chart
# --------------------------------
plt.subplot(2, 2, 1)  # (rows, columns, position)
months = ['Jan', 'Feb', 'Mar', 'Apr']
sales = [50, 60, 70, 80]
plt.plot(months, sales)
plt.title("Line Chart")

# --------------------------------
# 2. Bar Chart
# --------------------------------
plt.subplot(2, 2, 2)
courses = ['Python', 'Java', 'DS', 'Web']
students = [40, 30, 25, 35]
plt.bar(courses, students)
plt.title("Bar Chart")

# --------------------------------
# 3. Pie Chart
# --------------------------------
plt.subplot(2, 2, 3)
labels = ['A', 'B', 'C', 'D']
sizes = [30, 25, 20, 25]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")

# --------------------------------
# 4. Scatter Plot
# --------------------------------
plt.subplot(2, 2, 4)
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
plt.scatter(x, y)
plt.title("Scatter Plot")

# Adjust layout to prevent overlap
plt.tight_layout()

# Display all subplots
plt.show()
```
