```python
#Q1.Marks analysis with condition
import pandas as pd

marks = [45, 67, 89, 34, 90, 55, 72]
subjects = ['Math', 'Physics', 'Chemistry', 'English', 'CS', 'Biology', 'History']

s = pd.Series(marks, index=subjects)

print("Original Series:\n", s)

# Task 1: Find subjects with marks > 70
subjects_above_70 = s[s > 70]
print("\nSubjects with marks > 70:\n", subjects_above_70)

# Task 2: Replace marks < 40 with "Fail"
s_replaced = s.copy().astype(object) # Explicitly cast to object dtype to handle mixed types
s_replaced[s_replaced < 40] = "Fail"
print("\nSeries after replacing marks < 40 with 'Fail':\n", s_replaced)

# Task 3: Count how many subjects scored above average
average_mark = s.mean()
subjects_above_average_count = (s > average_mark).sum()
print(f"\nAverage mark: {average_mark:.2f}")
print(f"Number of subjects scored above average: {subjects_above_average_count}")
```

```python
#Q2.salary increament logic
import pandas as pd

salary = [25000, 32000, 28000, 40000, 50000]
emp = ['A', 'B', 'C', 'D', 'E']

s = pd.Series(salary, index=emp)

print("Original Salaries:\n", s)

# Task 1: Increase salary by 10% where salary < 30000
s_after_increase = s.copy().astype(float) # Explicitly cast to float to handle percentage increases
s_after_increase[s_after_increase < 30000] *= 1.10
print("\nSalaries after 10% increase for < 30000:\n", s_after_increase)

# Task 2: Decrease salary by 5% where salary > 45000
s_final = s_after_increase.copy()
s_final[s_final > 45000] *= 0.95
print("\nSalaries after 5% decrease for > 45000:\n", s_final)

# Task 3: Find total salary expense after changes
total_salary_expense = s_final.sum()
print(f"\nTotal salary expense after all changes: {total_salary_expense:.2f}")
```

```python
#Q3. Handling Missing Data Smartly
import pandas as pd
import numpy as np

data = [10, None, 30, None, 50, 60, None]
s = pd.Series(data)

print("Original Series with missing values:\n", s)

# Task 1: Count missing values before filling
missing_before = s.isnull().sum()
print(f"\nMissing values before filling: {missing_before}")

# Calculate mean for filling
series_mean = s.mean()
print(f"Series mean for filling: {series_mean:.2f}")

# Task 2: Fill missing values with mean
s_filled = s.fillna(series_mean)
print("\nSeries after filling missing values with mean:\n", s_filled)

# Task 3: Count missing values after filling
missing_after_fill = s_filled.isnull().sum()
print(f"Missing values after filling: {missing_after_fill}")

# Task 4: Drop remaining nulls (if any)
s_final_cleaned = s_filled.dropna()
print("\nSeries after dropping any remaining nulls (if any):\n", s_final_cleaned)

# Count missing values after dropping (should be 0)
missing_after_drop = s_final_cleaned.isnull().sum()
print(f"Missing values after dropping: {missing_after_drop}")
```

```python
#Q4. Student Grades System
import pandas as pd

marks = [95, 82, 67, 58, 73, 49, 88]
students = ['S1','S2','S3','S4','S5','S6','S7']

s = pd.Series(marks, index=students)

print("Original Marks Series:\n", s)

# Function to assign grades
def assign_grade(mark):
    if mark >= 90:
        return 'A'
    elif mark >= 75:
        return 'B'
    elif mark >= 60:
        return 'C'
    else:
        return 'D'

# Task 1: Assign grades using apply()
grades = s.apply(assign_grade)
print("\nAssigned Grades:\n", grades)

# Task 2: Count number of students in each grade
grade_counts = grades.value_counts().sort_index()
print("\nNumber of students in each grade:\n", grade_counts)
```

```python
#Q5. Series Alignment
import pandas as pd

s1 = pd.Series([10, 20, 30], index=['A', 'B', 'C'])
s2 = pd.Series([5, 15, 25], index=['B', 'C', 'D'])

print("Series 1:\n", s1)
print("Series 2:\n", s2)

# Task 1: Add two Series with different indices
s_add = s1 + s2
print("\nSum of Series (s1 + s2):\n", s_add)

# Task 2: Replace NaN with 0 and recompute sum
s_add_filled = s1.add(s2, fill_value=0)
print("\nSum of Series after filling NaN with 0 (s1.add(s2, fill_value=0)):\n", s_add_filled)
```

```python
#Q6. Ranking System
import pandas as pd

scores = [88, 92, 79, 93, 85]
players = ['P1','P2','P3','P4','P5']

s = pd.Series(scores, index=players)

print("Original Scores Series:\n", s)

# Task 1: Rank players based on scores (higher score means lower rank number)
ranks = s.rank(ascending=False, method='min') # 'min' method assigns the minimum rank to tied scores
print("\nPlayer Ranks (1=highest score):\n", ranks)

# Task 2: Display top 3 players
# Sort the series in descending order and take the top 3
top_3_players = s.sort_values(ascending=False).head(3)
print("\nTop 3 Players:\n", top_3_players)

# Task 3: Find player with lowest score
lowest_score_player = s.idxmin()
lowest_score = s.min()
print(f"\nPlayer with the lowest score: {lowest_score_player} (Score: {lowest_score})")
```

```python
#Q7. String Operations in Series
import pandas as pd

names = ['anuj', 'rahul', 'sneha', 'kiran', 'amit']
s = pd.Series(names)

print("Original Series:\n", s)

# Task 1: Convert all names to uppercase
s_upper = s.str.upper()
print("\nNames in uppercase:\n", s_upper)

# Task 2: Find names containing letter 'a'
# Using .str.contains() with case=False to find 'a' or 'A'
names_with_a = s[s.str.contains('a', case=False)]
print("\nNames containing letter 'a':\n", names_with_a)

# Task 3: Replace 'anuj' with 'Anuj Kumar'
# Using .str.replace() to find and replace the specific string
s_replaced = s.str.replace('anuj', 'Anuj Kumar')
print("\nSeries after replacing 'anuj' with 'Anuj Kumar':\n", s_replaced)
```

```python
#Q8. Chained Operations
import pandas as pd
import numpy as np

data = [5, None, 15, 25, None, 35, 45]
s = pd.Series(data)

print("Original Series with missing values:\n", s)

# Task 1: Fill missing values with median
median_value = s.median()
s_filled = s.fillna(median_value)
print(f"\nSeries after filling missing values with median ({median_value}):\n", s_filled)

# Task 2: Filter values greater than 20
s_filtered = s_filled[s_filled > 20]
print("\nSeries after filtering values greater than 20:\n", s_filtered)

# Task 3: Sort in descending order
s_sorted = s_filtered.sort_values(ascending=False)
print("\nSeries after sorting in descending order:\n", s_sorted)

# Task 4: Reset index
s_final = s_sorted.reset_index(drop=True)
print("\nSeries after resetting index:\n", s_final)
```

```python
#Q9.sales parformance:
import pandas as pd

sales = [200, 450, 300, 150, 500, 700, 100]
days = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']

s = pd.Series(sales, index=days)

print("Original Sales Series:\n", s)

# Task 1: Find best and worst sales day
best_sales_day = s.idxmax()
worst_sales_day = s.idxmin()
best_sales_value = s.max()
worst_sales_value = s.min()
print(f"\nBest Sales Day: {best_sales_day} (Sales: {best_sales_value})")
print(f"Worst Sales Day: {worst_sales_day} (Sales: {worst_sales_value})")

# Task 2: Calculate total and average sales
total_sales = s.sum()
average_sales = s.mean()
print(f"\nTotal Sales: {total_sales:.2f}")
print(f"Average Sales: {average_sales:.2f}")

# Task 3: Apply 20% bonus on days where sales > 400
s_bonused = s.copy().astype(float) # Ensure float dtype for calculations
s_bonused[s_bonused > 400] *= 1.20
print("\nSales after applying 20% bonus for sales > 400:\n", s_bonused)

# Task 4: Sort sales in descending order
s_sorted_desc = s.sort_values(ascending=False)
print("\nSales sorted in descending order:\n", s_sorted_desc)
```
